```python
import pandas as pd
import re
import unicodedata
import nltk
nltk.download('punkt')
from nltk.tokenize import sent_tokenize

# === 1. Load your original file ===
df = pd.read_csv('/content/islamweb_fatwas_qna.csv')  # Update path

# === 2. Define preprocessing functions ===
def remove_diacritics(text):
    arabic_diacritics = re.compile("""
                             ّ    | # Shadda
                             َ    | # Fatha
                             ً    | # Tanwin Fath
                             ُ    | # Damma
                             ٌ    | # Tanwin Damm
                             ِ    | # Kasra
                             ٍ    | # Tanwin Kasr
                             ْ    | # Sukun
                             ـ     # Tatwil/Kashida
                         """, re.VERBOSE)
    return re.sub(arabic_diacritics, '', text)

def remove_opening(text):
    patterns = [
        r"^الحمد لله.*?أما بعد[^\:]*:",
        r"^الحمد لله.*?أما بعد",
        r"^الحمد لله.*?، وبعد",
    ]
    for pattern in patterns:
        text = re.sub(pattern, '', text, flags=re.DOTALL)
    return text.strip()

def clean_text(text):
    text = re.sub(r'\s+', ' ', text)  # Remove extra spaces
    text = text.strip()
    return text

# === 3. Apply cleaning ===
rows = []
for idx, row in df.iterrows():
    question = clean_text(row['question'])
    context = row['answer']

    # Skip if context is empty
    if pd.isna(context) or len(context.strip()) == 0:
        continue

    # Clean context
    context = remove_diacritics(context)
    context = remove_opening(context)
    context = clean_text(context)

    # Skip if context becomes meaningless
    if len(context) < 30 or context in ['والله أعلم.', '']:
        continue

    # Assume full context is the answer (for now)
    answer_text = context
    answer_start = context.find(answer_text)  # Always 0 in this assumption

    rows.append({
        'text': context,
        'question': question,
        'answer_start': answer_start,
        'answer_text': answer_text
    })

# === 4. Create new dataframe ===
qa_df = pd.DataFrame(rows)

# === 5. Save ready-to-train dataset ===
qa_df.to_csv('islamweb_fatwas_QA_ready.csv', index=False, encoding='utf-8-sig')

print(f"✅ Dataset prepared with {len(qa_df)} QA pairs and saved as 'islamweb_fatwas_QA_ready.csv'.")
```

```python
# === CONTINUING AFTER YOUR SEGMENT ===

from datasets import load_dataset, Dataset
from transformers import TrainingArguments, Trainer
import pandas as pd

# === 1. Load your dataset (your clean_QA_dataset.csv) ===
df = pd.read_csv('/content/clean_QA_dataset_ready.csv')  # Update path if needed

# Transform into Huggingface Dataset
dataset = Dataset.from_pandas(df)

# === 2. Preprocessing Function ===
def prepare_features(examples):
    tokenized = tokenizer(
        examples["question"],
        examples["text"],
        padding="max_length",
        truncation=True,
        max_length=128,
        return_offsets_mapping=True
    )

    tokenized["start_positions"] = []
    tokenized["end_positions"] = []

    for i in range(len(tokenized["input_ids"])):
        offsets = tokenized["offset_mapping"][i]
        input_ids = tokenized["input_ids"][i]
        cls_index = input_ids.index(tokenizer.cls_token_id)

        sequence_ids = tokenized.sequence_ids(i)
        answer = examples["answer_text"][i]
        start_char = examples["answer_start"][i]
        end_char = start_char + len(answer)

        token_start_index = 0
        while sequence_ids[token_start_index] != 1:
            token_start_index += 1

        token_end_index = len(input_ids) - 1
        while sequence_ids[token_end_index] != 1:
            token_end_index -= 1

        if not (start_char <= end_char):
            tokenized["start_positions"].append(cls_index)
            tokenized["end_positions"].append(cls_index)
        else:
            while token_start_index < len(offsets) and offsets[token_start_index][0] <= start_char:
                token_start_index += 1
            while offsets[token_end_index][1] >= end_char:
                token_end_index -= 1
            tokenized["start_positions"].append(token_start_index - 1)
            tokenized["end_positions"].append(token_end_index + 1)

    return tokenized

# Apply preprocessing
tokenized_dataset = dataset.map(prepare_features, batched=True, remove_columns=dataset.column_names)

# === 3. Define Training Arguments ===
training_args = TrainingArguments(
    output_dir="./qa-bert-arabic",
    do_train=True,
    do_eval=True,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    learning_rate=2e-5,
    num_train_epochs=3,
    weight_decay=0.01,
    warmup_steps=100,
    logging_dir="./logs",
    logging_steps=100,
    save_steps=500,
    save_total_limit=2,
    fp16=True
)

# === 4. Define Trainer ===
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset,
    eval_dataset=tokenized_dataset,  # In reality, split into train/test later!
    tokenizer=tokenizer
)

# === 5. Start Fine-tuning! ===
trainer.train()
```

```python
# === 0. Install packages first (if not yet) ===
# pip install rank_bm25
# pip install sentence-transformers
# pip install faiss-cpu

import pandas as pd
import faiss
from rank_bm25 import BM25Okapi
from sentence_transformers import SentenceTransformer
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize

# === 1. Load your dataset ===
df = pd.read_csv('/content/clean_QA_dataset_ready.csv')
contexts = df['text'].tolist()

# === 2. Prepare BM25 index ===
tokenized_contexts = [word_tokenize(doc) for doc in contexts]
bm25 = BM25Okapi(tokenized_contexts)

# === 3. Prepare Dense Embedding model and FAISS index ===
dense_model = SentenceTransformer('sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2')
context_embeddings = dense_model.encode(contexts, show_progress_bar=True, batch_size=32)

embedding_size = context_embeddings.shape[1]
faiss_index = faiss.IndexFlatL2(embedding_size)
faiss_index.add(context_embeddings)

print(f"✅ FAISS index built with {faiss_index.ntotal} fatwas.")

# === 4. Define keyword list related to Tawaf, shoes, mosque etiquette ===
important_keywords = ["طواف", "سعي", "مسجد", "صلاة", "نعال", "حذاء", "خروج", "دخول المسجد", "طهارة", "قذارة"]

# === 5. Improved Retrieval Function ===
def improved_hybrid_retrieve(question, bm25_top_k=20, final_top_k=3):
    # Step 1: Expand query tokens
    tokenized_query = word_tokenize(question)
    expanded_query = tokenized_query + important_keywords  # Boost important keywords

    # Step 2: BM25 initial retrieval
    bm25_scores = bm25.get_scores(expanded_query)
    bm25_top_indices = sorted(range(len(bm25_scores)), key=lambda i: bm25_scores[i], reverse=True)[:bm25_top_k]
    bm25_top_contexts = [(contexts[i], i) for i in bm25_top_indices]

    # Step 3: Hard Filter BM25 candidates: keep only fatwas containing important keywords
    filtered_candidates = []
    for text, idx in bm25_top_contexts:
        if any(kw in text for kw in important_keywords):
            filtered_candidates.append((text, idx))

    # If filtering removed all candidates, fallback to original bm25 candidates
    if not filtered_candidates:
        filtered_candidates = bm25_top_contexts

    candidate_texts = [t for (t, idx) in filtered_candidates]
    candidate_indices = [idx for (t, idx) in filtered_candidates]

    # Step 4: Dense re-ranking
    candidate_embeddings = dense_model.encode(candidate_texts)
    question_embedding = dense_model.encode([question])

    distances = ((candidate_embeddings - question_embedding) ** 2).sum(axis=1)
    reranked_indices = distances.argsort()[:final_top_k]

    results = []
    for idx in reranked_indices:
        fatwa_text = candidate_texts[idx]
        fatwa_distance = distances[idx]
        results.append((fatwa_text, fatwa_distance))

    return results

# === 6. Example usage ===
query = "هل يجوز للمقتدر أخذ كفن من المسجد؟ وهل يجب عليه رد مثله لاحقا؟ وكيف يتصرف القائم على المسجد إذا طلب المقتدر كفنا؟ وجزاكم الله خيرا"
retrieved_fatwas = improved_hybrid_retrieve(query, bm25_top_k=20, final_top_k=3)

for i, (fatwa, distance) in enumerate(retrieved_fatwas):
    print(f"\nRank {i+1} | Distance: {distance:.4f}")
    print(fatwa)
```

### **5. ANSWER GENERATION**

```python
# === Install Transformers ===
# pip install transformers datasets

from datasets import Dataset
import pandas as pd
from transformers import AutoTokenizer, AutoModelForQuestionAnswering, TrainingArguments, Trainer

# === 1. Load your dataset ===
df = pd.read_csv('/content/clean_QA_dataset_ready.csv')

# Huggingface Datasets needs a Dataset object
dataset = Dataset.from_pandas(df)

# === 2. Load Pretrained BERT Model ===
model_name = "asafaya/bert-base-arabic"  # Arabic BERT
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForQuestionAnswering.from_pretrained(model_name)

# === 3. Preprocessing function ===
def prepare_train_features(examples):
    tokenized = tokenizer(
        examples['question'],
        examples['text'],
        truncation=True,          # Fix here (instead of "only_second")
        padding="max_length",
        max_length=256,            # Reduce to 256 (adapted to short fatwas)
        return_offsets_mapping=True
    )

    offset_mapping = tokenized.pop("offset_mapping")
    tokenized["start_positions"] = []
    tokenized["end_positions"] = []

    for i, offsets in enumerate(offset_mapping):
        start_char = examples['answer_start'][i]
        end_char = start_char + len(examples['answer_text'][i])

        sequence_ids = tokenized.sequence_ids(i)
        context_start = sequence_ids.index(1)
        context_end = len(sequence_ids) - sequence_ids[::-1].index(1) - 1

        token_start = context_start
        while token_start <= context_end and offsets[token_start][0] <= start_char:
            token_start += 1

        token_end = context_end
        while token_end >= context_start and offsets[token_end][1] >= end_char:
            token_end -= 1

        tokenized["start_positions"].append(token_start - 1)
        tokenized["end_positions"].append(token_end + 1)

    return tokenized

# === 4. Preprocess Dataset ===
tokenized_dataset = dataset.map(prepare_train_features, batched=True, remove_columns=dataset.column_names)

# === 5. Define TrainingArguments ===
from transformers import TrainingArguments

training_args = TrainingArguments(
    output_dir="./qa-bert-fatwa",
    do_train=True,
    do_eval=False,  # no evaluation during training
    per_device_train_batch_size=8,
    learning_rate=2e-5,
    num_train_epochs=3,
    save_steps=500,
    save_total_limit=2,
    weight_decay=0.01,
    logging_dir="./logs",
    logging_steps=100,
    fp16=True  # use mixed precision if available
)


# === 6. Define Trainer ===
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset,
    tokenizer=tokenizer
)

# === 7. Start Fine-tuning ===
trainer.train()
```

```python
# === Save both model and tokenizer properly ===
trainer.save_model("/content/qa-bert-fatwa")        # saves model
tokenizer.save_pretrained("/content/qa-bert-fatwa") # saves tokenizer files
```

```python
# === Install if missing ===
# pip install gradio faiss-cpu

import gradio as gr
from transformers import AutoTokenizer, AutoModelForQuestionAnswering, AutoModel
import torch
import faiss
import numpy as np
import pandas as pd

# === 1. Load your fine-tuned QA model ===
model_path = "/content/qa-bert-fatwa"
qa_tokenizer = AutoTokenizer.from_pretrained(model_path)
qa_model = AutoModelForQuestionAnswering.from_pretrained(model_path)

# === 2. Load your fatwa dataset ===
df = pd.read_csv('/content/clean_QA_dataset_ready.csv')

# Important: we need both questions and texts
questions = df['question'].tolist()
contexts = df['text'].tolist()

# === 3. Use Stronger Arabic Retriever
retriever_tokenizer = AutoTokenizer.from_pretrained("aubmindlab/bert-base-arabertv2")
retriever_model = AutoModel.from_pretrained("aubmindlab/bert-base-arabertv2")

# === 4. Build retrieval embeddings (on QUESTIONS not texts!)
def embed(texts):
    encoded_input = retriever_tokenizer(texts, padding=True, truncation=True, return_tensors="pt", max_length=512)
    with torch.no_grad():
        model_output = retriever_model(**encoded_input)

    attention_mask = encoded_input['attention_mask']
    token_embeddings = model_output.last_hidden_state
    input_mask_expanded = attention_mask.unsqueeze(-1).expand(token_embeddings.size()).float()

    sum_embeddings = torch.sum(token_embeddings * input_mask_expanded, 1)
    sum_mask = torch.clamp(input_mask_expanded.sum(1), min=1e-9)
    mean_embeddings = sum_embeddings / sum_mask

    return mean_embeddings.cpu().numpy()

# Embed the questions now
question_embeddings = embed(questions)

# Build FAISS index
dimension = question_embeddings.shape[1]
index = faiss.IndexFlatL2(dimension)
index.add(question_embeddings)

# === 5. Define new search and QA functions ===
def search_best_fatwa(user_question, k=3):
    user_embedding = embed([user_question])
    distances, indices = index.search(user_embedding, k)

    best_idx = indices[0][0]
    best_context = contexts[best_idx]
    return best_context

def answer_fatwa(user_question):
    # 1. Search best matching fatwa
    context = search_best_fatwa(user_question, k=3)

    # 2. Extract answer
    inputs = qa_tokenizer(user_question, context, return_tensors="pt", truncation=True, max_length=512)
    with torch.no_grad():
        outputs = qa_model(**inputs)

    answer_start = torch.argmax(outputs.start_logits)
    answer_end = torch.argmax(outputs.end_logits) + 1

    answer = qa_tokenizer.convert_tokens_to_string(
        qa_tokenizer.convert_ids_to_tokens(inputs["input_ids"][0][answer_start:answer_end])
    ).strip()

    if len(answer) < 3:
        return "❗ لم يتم العثور على إجابة دقيقة لهذا السؤال من قاعدة البيانات."

    return f"❓ السؤال:\n{user_question}\n\n✅ الإجابة:\n{answer}"

# === 6. Create Gradio Interface ===
iface = gr.Interface(
    fn=answer_fatwa,
    inputs=gr.Textbox(lines=2, placeholder="اكتب سؤالك هنا...", label="السؤال"),
    outputs="text",
    title="نظام الإجابة على الفتاوى الإسلامية (Islamic Fatwa QA System)",
    description="ادخل سؤالك، وسيقوم النظام بالبحث عن أقرب فتوى واستخراج الإجابة بدقة باستخدام الذكاء الاصطناعي."
)

# === 7. Launch Gradio App ===
iface.launch(share=True)
```
